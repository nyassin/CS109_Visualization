###GCMap Visualization

To develop a better understanding of flight routes, we used GCMap to visualize flights between the top 20 airports. Longtitude and Latitude data points are from OpenFlight database. Results are in files All_Visualization.png and Delays_Visualization.png. 


```python
#load up airport coordinates to display using GCMAP
#load smalldf.csv, which is a representative sample of our 3.5 million flights.
data =pd.read_csv("smalldf.csv")

#initialize arrays
lon_origin = []
lat_origin = []
lat_destination = []
lon_destination = []

for count, row in data.iterrows():
    destination = row['Dest']
    origin = row['Origin']

    # if destination in airports_coordinates and row['DEP_DELAY'] > 15:
    if destination in airports_coordinates:
       
        #append coordinates of origin
        lon_origin.append(airports_coordinates[origin]['lon'])
        lat_origin.append(airports_coordinates[origin]['lat'])
       
        #append coordinates of destination
        lon_destination.append(airports_coordinates[destination]['lon'])
        lat_destination.append(airports_coordinates[destination]['lat'])
   
```

```python
from gcmap import GCMapper

#initialize GCMapper with high resolution output
gcm = GCMapper(width=6000, height=6000, gc_resolution=1000)
gcm.set_data(lon_origin, lat_origin, lon_destination, lat_destination)
img = gcm.draw()
img.save('All_Visualization.png')
```

###D3 Visualizations

We were not satisfied with results generated from GCMap. They did not properly show the U.S map and it was hard to gain any real understanding from the two files. We want to make flight visualizations more interactive to heopfully gain insight by simply looking at the visualization. 

D3 is a powerful Javascript library that helps visualize data. This section of code generates CSV lists of data that we later on plug into D3 visualizations. Essentially, we are keeping the columns that are most interesting in a convenient CSV format.  


```python
import csv
data =pd.read_csv("representative_sample_of_flights.csv")
c = csv.writer(open("visualizationData.json", "wb"))

#configure the headers
c.writerow(["date", "delay", "distance", "temp", "visibility", "windspeed", "pressure", "origin", "destination"])

P_mean = data.Pressure.mean()

for count, row in data.iterrows():
    destination = row['Dest']
    origin = row['Origin']
    
    #make Datetime format compatible with the format in the visualization
    datetimeFormat = datetime.datetime.strptime( row['DateTime'], '%Y-%m-%d %H:%M:%S')
    date = str(datetimeFormat.strftime('%m'))+ str(datetimeFormat.strftime('%d')) + str(datetimeFormat.strftime('%H')) + str(datetimeFormat.strftime('%M'))
    
    #make sure there are no NaNs and normalize delay minute time.
    if np.isnan(row['Delay_Minutes']):
        delay = 0
    else:
        delay = int(row['Delay_Minutes'])
    
    if delay < 0:
        delay = 0
    
    #write row to CSV file and save it
    json_row = [date, delay, int(row['Distance']), row['Temp'], row['Visibility'], row['Windspeed'], row['Pressure'] - P_mean, origin, destination]
    c.writerow(json_row)

```
