##Two Types of Delay Variables We Are Interested In
###NAS_Delay
Nas_Delay is described as avoidable delays that with a better run system would be worked out. This column captures the idea of a mild weather delay that would be avoidable with better preparation on the part of the airport or FAA.
###Weather_Delay
Weather Delays are delays due to extreeme weather and would not be avoidable.

####Predictable Delays
One problem with flight delays is that many of them are due to unpredictable random events such as mechanical failure. We hope by limiting ourselves to predicting delays greater than 15min that are related to weather, we can actually make meaningful predictions. We realized after running logistic regresion that it is too difficult to predict all delays, so we limit ourselves to predicting those related to weather (NAS_Delay and Weather_Delay). 

####Cancelled Flights
There are also a small number of flights in the dataset that are cancelled altogether. We decided to include these as delays related to weather, because the weather for cancelled flights is more extreme than the weather for most flights (this is explored at the bottom of the notebook).

```python
def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
def pretty_bars(limit_num, names, values, xlab, ylab, title):
    graph_subset_names = list(names[:limit_num])
    graph_subset_values = list(values[:limit_num])
    
    fig = plt.figure()
    fig.set_size_inches(15,9)
    ax = fig.add_subplot(111)
    ind = np.arange(limit_num)
    w = 0.7
    
    ax.bar(ind, graph_subset_values, width = w)#, align='center')
    ax.set_xticks(ind)
    ax.set_xticklabels(graph_subset_names, rotation=90,fontsize=8)
    
    plt.xlabel(xlab)
    plt.ylabel( ylab)
    plt.title(title)
    remove_border()

LIMIT_NUM_AIRPORTS = 100 # max val = 306

pretty_bars(LIMIT_NUM_AIRPORTS, origin_airport_sorted, num_depart_flights_sorted,
            "Airport", "Number of Departures", 
            "Number of Departures per Airport for Top {0} Airports".format(LIMIT_NUM_AIRPORTS))
```

### What Percentage of Flights are Delayed?

```python
df['Valid_Delay_Time'] = df['Delay_Time'].map(lambda x: x >= 0)
#print df.head(10)

DELAY_THRESHOLD = 15
#Add a new column "Delayed" that is a bool if it was delayed at all.
df['Delayed'] = df['Delay_Time'].map(lambda x: x >= DELAY_THRESHOLD)

group_valid_delay_time = df.groupby("Valid_Delay_Time")
valid_flights_df = pd.DataFrame()
invalid_delay_time = pd.DataFrame()
for key,value in group_valid_delay_time:
    if not key:
        invalid_delay_time = value
    else:
        valid_flights_df = value

number_invalid_delay_time = len(invalid_delay_time.index)
valid_flights = len(valid_flights_df.index)
print "There are {0:,} flights with Nan for their Departure Time and {1:,} flights with valid Departure Times".format(number_invalid_delay_time,
                                                                                                                      valid_flights)
percent_invalid_departure = (float(number_invalid_delay_time) / (number_invalid_delay_time + valid_flights)) * 100
print "In January 2013, {0:0.2f} percent of flights had Nan for Departure time.".format(percent_invalid_departure)
```

```python
# Working only with the flights with valid delay times now
groupby_delay = valid_flights_df.groupby('Delayed')
delayed_frame = pd.DataFrame()
non_delayed_frame = pd.DataFrame()
for key,value in groupby_delay:
    if (key):
        delayed_frame = value
    else:
        non_delayed_frame=value
        

num_delayed = len(delayed_frame.index)
num_ontime = len(non_delayed_frame.index)
percent_delayed = float(num_delayed) / float(len(valid_flights_df.index)) *100
print "There are {0:,} flights that actually took off. Of those flights {1:,} were delayed more that {2} min which is {3:0.2f} percent.".format(len(valid_flights_df.index),
                                                                                                                    num_delayed,
                                                                                                                    DELAY_THRESHOLD,
                                                                                                                    percent_delayed)

```

```python
delayed_airport_groups = delayed_frame.groupby('ORIGIN')
delay_origin_airport = []
num_delayed_flights = [] 
for key,value in delayed_airport_groups:
    delay_origin_airport.append(key)
    num_delayed_flights.append(len(value.index))

num_delayed_flights_sorted, delay_origin_airport_sorted = zip(*sorted(zip(num_delayed_flights, delay_origin_airport), reverse=True))

dict_num_departures = defaultdict(lambda: float('Nan'))
for i in range(len(num_delayed_flights_sorted)):
    dict_num_departures[origin_airport_sorted[i]] = num_depart_flights_sorted[i]
    
dict_delay_num_departures = defaultdict(lambda: 0.0)
for i in range(len(num_delayed_flights_sorted)):
    dict_delay_num_departures[delay_origin_airport_sorted[i]] = num_delayed_flights_sorted[i]
    
percent_delayed = []
percent_delayed_airport = []
for airport in delay_origin_airport:
    if not math.isnan( dict_num_departures[airport]):
        percent_delayed_airport.append(airport)
        percent_delay = float(dict_delay_num_departures[airport])/float(dict_num_departures[airport]) *100
        percent_delayed.append(percent_delay)
    
percent_delayed_sorted, percent_delayed_airport_sorted = zip(*sorted(zip(percent_delayed, percent_delayed_airport), reverse=True))

print "Airports with most delays per flights" 
print"(from all airports)"
for i in range(20):
    print "{0:2}. {1} {2:.2f}".format(i+1, percent_delayed_airport_sorted[i], percent_delayed_sorted[i])

top_20_delays = []
top_20_delays_airports = [] 
for airport in twenty_busiest:
    top_20_delays_airports.append(airport)
    percent = float(dict_delay_num_departures[airport])/float(dict_num_departures[airport]) *100
    top_20_delays.append(percent)
top_20_delays_sorted, top_20_delays_airports_sorted = zip(*sorted(zip(top_20_delays, top_20_delays_airports), reverse=True))

print 
print "The Twenty Airports with the most Delayed Departures in January 2013\n"
print " "*10+"Delayed Departures" + " "*12 + "Top 20 Airports for Comparison" + " "*5 + "Percent of Flights Delayed"
print " "*80 + " for 20 Biggest"
for i in range(20):
    #print "{0:2}. {1} with {2:6,} Delayed Flights".format((i +1),delay_origin_airport_sorted[i],num_delayed_flights_sorted[i]) 
    print "{0:2}. {1} with {2:6,} Delayed Flights     {0:2}. {3} with {4:6,} departures    {0:2}. {5} {6:0.2f} percent delayed ".format((i +1),
                                                                                                 delay_origin_airport_sorted[i],
                                                                                                 num_delayed_flights_sorted[i],
                                                                                                 origin_airport_sorted[i] , 
                                                                                                 num_depart_flights_sorted[i], 
                                                                                                 top_20_delays_airports_sorted[i], 
                                                                                                 top_20_delays_sorted[i]) 
    
 
```

Below we look at how day of the week affects delays.

```python
Day_of_week = {1:'Monday',2:'Tuesday',
               3:'Wednesday',4:'Thursday',
               5:'Friday',6:'Saturday',
               7:'Sunday',9:'unknown'}

group_dow = big_df.groupby('Day_of_week')

Days = []
flights = []
delayed_flights=[]

for key, value in group_dow:
    flights.append(len(value.index))
    delayed_flights.append(sum(value.Delayed))
    Days.append(key)
    
```
