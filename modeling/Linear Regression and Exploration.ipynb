# Linear Regression and Exploration
Here we look at some simple linear regressions between delay time and different weather variables to get a feel for the dataset.

```python
%matplotlib inline
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
df = pd.read_csv('/Users/jamielaw-smith/Dropbox/Project/maindata/fulldf.csv')
df.columns
```

When Delay_Minutes is a NaN, this means the flight was cancelled. When we are only classifying flights in the binary of Delayed/Not Delayed (depending on whether they are delayed more than 15 minutes), we would just count these flights as delayed. But when using Delay_Minutes, which is continuous, we will set these values to the max delay time.

```python
# what percentage of flights is cancelled?
perc_canc = float(len(df[np.isnan(df.Delay_Minutes)]))/float(len(df))
print "Fraction of flights cancelled is %.4f" % perc_canc

# set NaN values of Delay_Minutes to max delayed time
df.Delay_Minutes[np.isnan(df.Delay_Minutes)] = df.Delay_Minutes.max()
```

## Scatter plots and linear regression

First remove rows from df that have NaNs in the weather variables we are interested in. These only comprise a small percentage of our total rows.

```python
# Airtime is not a weather variable, but it is important in predicting delays
weather_cols = ['Temp', 'Windspeed', 'Visibility', 'Pressure', 'Dewpoint', 'Airtime']

for col in weather_cols:
    print "{0:.4f}% of {1} rows are NaNs".format(len(df[np.isnan(df[col])])*100/float(len(df)), col)
    df = df[np.logical_not(np.isnan(df[col]))]
```

Before making scatter plots, take a random subset 1/200th the size of the full dataframe (otherwise plots will be too crowded).

```python
subset_ixs = np.random.randint(0,len(df),len(df)/200)
subdf = df.iloc[subset_ixs]
```

Let's make some scatter plots of these different variables with minutes delayed, as well as the slope of the linear regression and other important metrics such as the r and p values.

```python
from scipy import stats

# make scatter plots
for col in weather_cols:
    plt.figure(figsize=(8,6))
    plt.scatter(subdf[col],subdf['Delay_Minutes'],marker='.', alpha=0.2)
    plt.xlabel(col)
    plt.ylabel('Delay_Minutes')
    remove_border()
    slope, intercept, r_value, p_value, std_err = stats.linregress(subdf[col],subdf['Delay_Minutes'])
    line = slope*subdf[col]+intercept
    plt.plot(subdf[col],line,'r-')
    title_str = 'Delay vs. {0}: slope = {1:.2f}, r-value = {2:.2f}, p-value = {3:.2f}, std_err = {4:.2f}'.format(col, slope, r_value, p_value, std_err)
    plt.title(title_str)
```

The p-values suggest that these measures are significant, but the actual slopes and r-values suggest that each measure does not have a huge effect on delays. Nonetheless, the slopes make sense: delays are slightly positively correlated with temperature and windspeed but negatively correlated with pressure and visibility.
